```python
import pandas as pd

df = pd.read_excel('47.xlsx', header=None)  # 'header=None' sütun adlarının olmadığını belirtir

df.columns = ['sirket_adi']  # Sütun adlarını burada belirtin

# İlk 500 satırı alalım
#df = df.iloc[18000:18500]

# Şirket isimlerini çekelim
company_names = df['sirket_adi']                 
```

```python
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

# Büyük sosyal medya ve istenmeyen siteleri hariç tutmak için liste tanımlayın
excluded_sites = ['facebook.com', 'linkedin.com', 'instagram.com', 'x.com', 'youtube.com', 
                  'find-and-update.company-information.service.gov.uk', 'wikipedia.org', 'github.com']

# Web sitelerini saklayacağımız bir liste oluştur
websites = []

# Arama işlemi için sayaç
counter = 0
total_sites_searched = 0  # Toplam site sayısını takip etmek için değişken

# 20 şirketlik gruplar halinde tarayıcıyı aç/kapat döngüsü
for company in company_names:
    # Tarayıcıyı her 10 aramadan sonra aç/kapat
    if counter % 10 == 0:
        if counter != 0:  # İlk işlemde kapat/bekle yapma
            driver.quit()  # Tarayıcıyı kapat
            print(f"{total_sites_searched} site tarandı, 10 saniye bekleniyor...")
            time.sleep(10)  # 10 saniye bekle

        # Yeni bir tarayıcı oturumu başlat
        driver = webdriver.Edge()
        
        # Google ana sayfasını bir kez aç
        driver.get("https://www.google.com")

    # Arama kutusunu bul
    search_box = driver.find_element(By.NAME, 'q')
    
    # Arama kutusunu temizle ve yeni sorguyu yazıp arat
    search_box.clear()
    search_query = f"{company} website"
    search_box.send_keys(search_query)
    search_box.send_keys(Keys.RETURN)

    # Sayfanın yüklenmesini bekle ve gecikme ekle
    time.sleep(random.uniform(2, 4))  # Rastgele 2-4 saniye bekler

    try:
        # Organik sonuçları bul (reklamları ve istenmeyen siteleri atlayarak)
        results = driver.find_elements(By.CSS_SELECTOR, 'div.yuRUbf a')

        # İlk geçerli sonucu bul veya bulunamadı mesajı ver, istenmeyen siteleri atla
        website = "Site bulunamadı"
        for result in results:
            url = result.get_attribute('href')
            
            # İstenmeyen siteleri kontrol et ve atla
            if any(excluded in url for excluded in excluded_sites):
                continue
            
            # İlk geçerli sonucu bul ve kaydet
            website = url
            break
        
    except Exception:
        website = "Site bulunamadı"

    # Web sitesi adresini listeye ekle
    websites.append(website)

    # Toplam site sayısını güncelle
    total_sites_searched += 1

    # Her arama döngüsü arasında rastgele bir gecikme ekle
    time.sleep(2)  # 1-3 saniye arası ek bekleme süresi

    # Sayaç arttır
    counter += 1

# Döngü tamamlandıktan sonra tarayıcıyı kapat
driver.quit()



# Web sitesi sonuçlarını bir DataFrame'e dönüştür
results_df = pd.DataFrame({
    'sirket_adi': company_names,
    'website': websites
})


# Dosya adı değişkenini tanımlayın (daha sonra kolayca değiştirilebilir)
csv_file_name = 'sirket_websiteleri_sonuclar23000_23666.csv'

# DataFrame'i CSV dosyasına kaydet
results_df.to_csv(csv_file_name, index=False, encoding='utf-8')
df_w = pd.read_csv('sirket_websiteleri_sonuclar23000_23666.csv')
df_w
```
